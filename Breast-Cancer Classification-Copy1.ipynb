LOADING THE BREAST CANCER DATASET 

```python
import pandas as pd  #importing pandas to work on breast cancer dataset
df=pd.read_csv("cancer.csv",header=None)       #reading the file in csv format
df
```

SUMMARIZING THE DATASET

REMOVING NULL VALUES FROM THE DATASET

```python
df1=df.dropna()
df1
```

LABELLING THE TARGET VALUES

```python
dict={"B":0,"M":1}
df1[1]=df1[1].map(dict)
df1[1].value_counts()
```

DIVIDING THE DATASET INTO INDEPENDENT VALUES AND TARGET VALUES

```python
x=df1.iloc[0:,:-1] 
print(x)
y=df1[1]
print(x.shape)
print(y.shape)
print(y)
```

DATA SPLITTING FOR TRAINING AND TESTING

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.model_selection import GridSearchCV
import numpy as np
```

IMPORTING MODELS

```python
from sklearn.linear_model import Perceptron
clf1=Perceptron()
from sklearn.linear_model import LogisticRegression
clf2=LogisticRegression()
from sklearn.tree import DecisionTreeClassifier
clf3=DecisionTreeClassifier()
```

#### Without Standardizing

MODEL 1: Perceptron

```python
clf1.fit(x_train,y_train)
y_test_pred1=clf1.predict(x_test)
y_train_pred1=clf1.predict(x_train)
train_acc1=accuracy_score(y_train,y_train_pred1)
test_acc1=accuracy_score(y_test,y_test_pred1)
print("Training Acc=",train_acc1)
print("Testing Acc=",test_acc1)
cr1=classification_report(y_test,y_test_pred1)
print(cr1)
cm1=confusion_matrix(y_test,y_test_pred1)
print(cm1)
sns.heatmap(cm1,annot=True,cbar=False)
```

MODEL 2: Logistic Regression

```python
clf2.fit(x_train,y_train)
y_test_pred2=clf2.predict(x_test)
y_train_pred2=clf2.predict(x_train)
train_acc2=accuracy_score(y_train,y_train_pred2)
test_acc2=accuracy_score(y_test,y_test_pred2)
print("Training Acc=",train_acc2)
print("Testing Acc=",test_acc2)
cr2=classification_report(y_test,y_test_pred2)
print(cr2)
cm2=confusion_matrix(y_test,y_test_pred2)
print(cm2)
sns.heatmap(cm2,annot=True,cbar=False)
```

MODEL 3: Decision Tree Classifier

```python
clf3.fit(x_train,y_train)
y_test_pred3=clf3.predict(x_test)
y_train_pred3=clf3.predict(x_train)
train_acc3=accuracy_score(y_train,y_train_pred3)
test_acc3=accuracy_score(y_test,y_test_pred3)
print("Training Acc=",train_acc3)
print("Testing Acc=",test_acc3)
cr3=classification_report(y_test,y_test_pred3)
print(cr3)
cm3=confusion_matrix(y_test,y_test_pred3)
print(cm3)
sns.heatmap(cm3,annot=True,cbar=False)
```

#### Standardizing

```python
sc.fit(x_train)
x_train_std=sc.transform(x_train)
x_test_std=sc.transform(x_test)
```

MODEL 1: Perceptron

```python
clf1.fit(x_train_std,y_train)
y_test_pred1s=clf1.predict(x_test_std)
y_train_pred1s=clf1.predict(x_train_std)
train_acc1s=accuracy_score(y_train,y_train_pred1s)
test_acc1s=accuracy_score(y_test,y_test_pred1s)
print("Training Acc=",train_acc1s)
print("Testing Acc=",test_acc1s)
cr1s=classification_report(y_test,y_test_pred1s)
print(cr1s)
cm1s=confusion_matrix(y_test,y_test_pred1s)
print(cm1s)
sns.heatmap(cm1s,annot=True,cbar=False)
```

MODEL 2: Logistic Regression

```python
clf2.fit(x_train_std,y_train)
y_test_pred2s=clf2.predict(x_test_std)
y_train_pred2s=clf2.predict(x_train_std)
train_acc2s=accuracy_score(y_train,y_train_pred2s)
test_acc2s=accuracy_score(y_test,y_test_pred2s)
print("Training Acc=",train_acc2s)
print("Testing Acc=",test_acc2s)
cr2s=classification_report(y_test,y_test_pred2s)
print(cr2s)
cm2s=confusion_matrix(y_test,y_test_pred2s)
print(cm2s)
sns.heatmap(cm2s,annot=True,cbar=False)
```

MODEL 3: Decision Tree Classifier

```python
clf3.fit(x_train_std,y_train)
y_test_pred3s=clf3.predict(x_test_std)
y_train_pred3s=clf3.predict(x_train_std)
train_acc3s=accuracy_score(y_train,y_train_pred3s)
test_acc3s=accuracy_score(y_test,y_test_pred3s)
print("Training Acc=",train_acc3s)
print("Testing Acc=",test_acc3s)
cr3s=classification_report(y_test,y_test_pred3s)
print(cr3s)
cm3s=confusion_matrix(y_test,y_test_pred3s)
print(cm3s)
sns.heatmap(cm3s,annot=True,cbar=False)
```

### With Hyper Parameter tuning

MODEL 1:Perceptron

```python
param_grid1 = {"tol":[1e-3],
               'max_iter':[4]
              } 
grid1 = GridSearchCV(clf1, param_grid1,cv=5)
grid1.fit(x_train, y_train)
y_test_pred1t=grid1.predict(x_test)
y_train_pred1t=grid1.predict(x_train)
train_acc1t=accuracy_score(y_train,y_train_pred1t)
test_acc1t=accuracy_score(y_test,y_test_pred1t)
print("Training Acc=",train_acc1t)
print("Testing Acc=",test_acc1t)
cr1t=classification_report(y_test,y_test_pred1t)
print(cr1t)
cm1t=confusion_matrix(y_test,y_test_pred1t)
print(cm1t)
sns.heatmap(cm1t,annot=True,cbar=False)
```

MODEL 2: Logistic Regression

```python
c_space=np.logspace(-5,8,15)
param_grid2 = {'C': c_space,
              "multi_class":['auto', 'ovr', 'multinomial']}
grid2 = GridSearchCV(clf2, param_grid2, cv=5)
grid2.fit(x_train, y_train)
y_test_pred2t=grid2.predict(x_test)
y_train_pred2t=grid2.predict(x_train)
train_acc2t=accuracy_score(y_train,y_train_pred2t)
test_acc2t=accuracy_score(y_test,y_test_pred2t)
print("Training Acc=",train_acc2t)
print("Testing Acc=",test_acc2t)
cr2t=classification_report(y_test,y_test_pred2t)
print(cr2t)
cm2t=confusion_matrix(y_test,y_test_pred2t)
print(cm2t)
sns.heatmap(cm2t,annot=True,cbar=False)
```

MODEL 3: Decision Tree Classifier

```python
c_space=np.logspace(-5,8,15)
param_grid3 = {"max_depth": [3, None],
              "criterion": ["gini", "entropy"]}
grid3 = GridSearchCV(clf3, param_grid3, cv=5)
grid3.fit(x_train, y_train)
y_test_pred3t=grid3.predict(x_test)
y_train_pred3t=grid3.predict(x_train)
train_acc3t=accuracy_score(y_train,y_train_pred3t)
test_acc3t=accuracy_score(y_test,y_test_pred3t)
print("Training Acc=",train_acc3t)
print("Testing Acc=",test_acc3t)
cr3t=classification_report(y_test,y_test_pred3t)
print(cr3t)
cm3t=confusion_matrix(y_test,y_test_pred3t)
print(cm3t)
sns.heatmap(cm3t,annot=True,cbar=False)
```

### Model Comparison
